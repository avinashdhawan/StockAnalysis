```python
#import dependancies
import pymongo
import pandas as pd

from glob import glob
import os
```

```python
#
stock_files = glob('../Components/*.csv')

for i, f in enumerate(stock_files):
    if i == 0:
        df = pd.read_csv(f)
        df['Stock'] = f
    else:
        tmp = pd.read_csv(f)
        tmp['Stock'] = f
        df = df.append(tmp)

df.head()
```

```python
df.Stock =[x.strip('"../Components\\"&''.csv') for x in df.Stock]
Stock_df = df[["Stock", "Date","Open","High","Low","Close","Adj Close","Volume"]]
```

```python
Stock_df
```

```python
# Connect to MongoDB

client = pymongo.MongoClient('localhost',27017)
db = client['stocks'] #access stocks db
```

```python
# create collection in stocks db
db.stock_data.drop()
collection = db['stock_data']
```

```python
# write df to dictionary
stock_dict = Stock_df.to_dict("records")
```

```python
# Insert collection
collection.insert_many(stock_dict)
```
